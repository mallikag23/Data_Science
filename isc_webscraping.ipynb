```python
import pandas as pd
import selenium
import time
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.select import Select
import undetected_chromedriver as uc
from selenium.webdriver.common.by import By

driver=webdriver.Chrome(r'F:\chromedriver_win32\chromedriver.exe')
```

```python
urls = []

base_url = "https://www.amazon.in/s?k=Sarees&page={}&qid=1632112444&ref=sr_pg_{}"

for i in range(2,10):
    
    driver.get(base_url.format(i,i))
    
    for url in driver.find_elements_by_xpath("//a[@class='a-link-normal s-no-outline']"):
        
        urls.append(url.get_attribute("href"))
```

```python
# Grabbing image source

images_sarees = []

for link in urls:
    
    driver.get(link)
    time.sleep(1)
    
    try:
        
        img = driver.find_element_by_css_selector("#landingImage")
    
        source= img.get_attribute('src')
    
        images_sarees.append(source)
        
    except NoSuchElementException:
        
        images_sarees.append("Image Not Found!")
```

```python
import requests
for i in range(len(images_sarees)):
    
    print("Downloading....")
    
    response= requests.get(images_sarees[i])
    
    file = open(r"E:\Data Trained_ Training\FlipRobo\Image Scraping and Classification Project\Images\Sarees\Sarees" +str(i)+".jpg", "wb")
    
    file.write(response.content)    
```
